```python
import pandas as pd

df = pd.read_csv("../../data/raw/transactions_data.csv", parse_dates=["date"])
```

```python
data = df[['date', 'amount', 'use_chip', 'mcc', 'errors']]
data
```

```python
import json

def traducir(x, dict):
    x = str(x)
    if x in dict:
        return dict[x]
    else:
        return 'Sin tipo'
        
def limpiar_error(e):
    if str(e) == 'nan':
        return 'Sin error'
    return e
        
def codificar_fecha(f):
    hora = f.hour
    minuto = f.minute
    return 60*hora+minuto

def madrugada(x):
    if x.hour < 7:
        return True
    return False
        


data['amount'] = data['amount'].map(lambda x: float(x[1:]))
with open('../../data/raw/mcc_codes.json', 'r') as file:
        dict_categorias = json.load(file)
data['mcc'] = data['mcc'].map(lambda x: traducir(x, dict_categorias))
data['errors'] = data['errors'].map(limpiar_error)
data['fecha_numerica'] = data['date'].map(codificar_fecha)
data['madrugada'] = data['date'].map(madrugada)
data['index'] = data.index.values
data
```

```python
data.to_csv('../../data/raw/train.csv', index=False)
```

```python
def encontrar_label(x,d):
    if str(x) in d['target']:
        return d['target'][str(x)]
    return 'No'

b = []
for i in train['index'].values:
    if str(i) in d['target']:
        b.append(True)
    else:
        b.append(False)

train_labeled = train[b]

train_labeled['label'] = train_labeled['index'].map(lambda x: encontrar_label(x, d))
train_labeled
```
