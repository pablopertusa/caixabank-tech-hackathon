```python
import pandas as pd

df = pd.read_csv("../../data/raw/transactions_data.csv", parse_dates=["date"])
```

```python
data = df[['date', 'amount', 'use_chip', 'mcc', 'errors']]
data
```

```python
import json

def traducir(x, dict):
    x = str(x)
    if x in dict:
        return dict[x]
    else:
        return 'Sin tipo'
        
def limpiar_error(e):
    if str(e) == 'nan':
        return 'Sin error'
    return e
        
def codificar_fecha(f):
    hora = f.hour
    minuto = f.minute
    return 60*hora+minuto

def madrugada(x):
    if x.hour < 7:
        return True
    return False
        


data['amount'] = data['amount'].map(lambda x: float(x[1:]))
with open('../../data/raw/mcc_codes.json', 'r') as file:
        dict_categorias = json.load(file)
data['mcc'] = data['mcc'].map(lambda x: traducir(x, dict_categorias))
data['errors'] = data['errors'].map(limpiar_error)
data['fecha_numerica'] = data['date'].map(codificar_fecha)
data['madrugada'] = data['date'].map(madrugada)
data['index'] = data.index.values
data
```

```python
from datetime import datetime
datetime.strptime('2019-10-31 23:56:00', '%Y-%m-%d %H:%M:%S').minute
```

```python
data.to_csv('../../data/raw/train.csv', index=False)
```

```python
import pandas as pd
train = pd.read_csv('../../data/raw/train.csv')
```

```python
train['index'].values
```

```python
import json
with open('../../data/raw/train_label_fraud.json', 'r') as f:
    d = json.load(f)
d
```

```python
def encontrar_label(x,d):
    if str(x) in d['target']:
        return d['target'][str(x)]
    return 'No'

b = []
for i in train['index'].values:
    if str(i) in d['target']:
        b.append(True)
    else:
        b.append(False)

train_labeled = train[b]

train_labeled['label'] = train_labeled['index'].map(lambda x: encontrar_label(x, d))
train_labeled
```

```python
train_labeled['label'].unique()
```

```python
train_labeled.to_csv('../../data/raw/train_labeled.csv', index=False)
```

```python
train_labeled.columns
```

```python
categorical_columns = ['use_chip', 'mcc', 'errors', 'madrugada']

df_encoded = train_labeled[['amount', 'use_chip', 'mcc', 'errors', 'fecha_numerica',
       'madrugada','label']]

df_encoded = pd.get_dummies(df_encoded, columns=categorical_columns)
df_encoded
```

```python
for col in df_encoded.columns:
    if col not in ['label']:
        df_encoded[col] = df_encoded[col].astype(float)

df_encoded
```

```python
df_encoded.to_csv('../../data/raw/train_num.csv', index=False)
```
